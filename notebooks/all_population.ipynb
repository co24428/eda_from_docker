```python
li1 = ["5세", "6세", "7세"]
li2 = ["8세", "9세", "10세", "11세", "12세", "13세"]
li3 = ["14세", "15세", "16세"]
li4 = ["17세", "18세", "19세"]
li_all = li1 + li2 + li3 + li4
sum1_list = list()
sum2_list = list()
sum3_list = list()
sum4_list = list()

for one in state_list:
    tmp = test[ test["행정구역"] == one ]
    sum1 = 0
    sum2 = 0
    sum3 = 0
    sum4 = 0
    for i in li_all:
        if i in li1:
            sum1 = sum1 + int(tmp[ tmp["나이"] == i ]["value"])
        elif i in li2:
            sum2 = sum2 + int(tmp[ tmp["나이"] == i ]["value"])
        elif i in li3:
            sum3 = sum3 + int(tmp[ tmp["나이"] == i ]["value"])
        elif i in li4:
            sum4 = sum4 + int(tmp[ tmp["나이"] == i ]["value"])
        
    sum1_list.append(sum1)
    sum2_list.append(sum2)
    sum3_list.append(sum3)
    sum4_list.append(sum4)
kids_pop["유치원생"] = sum1_list
kids_pop["초등학생"] = sum2_list
kids_pop["중학생"] = sum3_list
kids_pop["고등학생"] = sum4_list
```

```python
test = pd.melt( all_pop.loc[:, ["행정구역", "나이", "총계"]], id_vars=['행정구역', '나이'] )
test.drop(index=test[test["나이"]=="계"].index, inplace=True) # 계인 인덱스 삭제
```

```python
state_list = test['행정구역'].unique()
sum0_list = list()
sum1_list = list()
sum2_list = list()
sum3_list = list()
sum4_list = list()
sum5_list = list()
sum6_list = list()
sum7_list = list()
sum8_list = list()
sum9_list = list()
sum10_list = list()

for one in state_list:
    tmp = test[ test["행정구역"] == one ]
    sum0 = 0
    sum1 = 0
    sum2 = 0
    sum3 = 0
    sum4 = 0
    sum5 = 0
    sum6 = 0
    sum7 = 0
    sum8 = 0
    sum9 = 0
    sum10 = 0
    
    for i in tmp.나이:
        if re.match(age0, i):
            sum0 = sum0 + int(tmp[ tmp["나이"] == i ]["value"])
        elif re.match(age10, i):
            sum1 = sum1 + int(tmp[ tmp["나이"] == i ]["value"])
        elif re.match(age20, i):
            sum2 = sum2 + int(tmp[ tmp["나이"] == i ]["value"])
        elif re.match(age30, i):
            sum3 = sum3 + int(tmp[ tmp["나이"] == i ]["value"])
        elif re.match(age40, i):
            sum4 = sum4 + int(tmp[ tmp["나이"] == i ]["value"])
        elif re.match(age50, i):
            sum5 = sum5 + int(tmp[ tmp["나이"] == i ]["value"])
        elif re.match(age60, i):
            sum6 = sum6 + int(tmp[ tmp["나이"] == i ]["value"])
        elif re.match(age70, i):
            sum7 = sum7 + int(tmp[ tmp["나이"] == i ]["value"])
        elif re.match(age80, i):
            sum8 = sum8 + int(tmp[ tmp["나이"] == i ]["value"])
        elif re.match(age90, i):
            sum9 = sum9 + int(tmp[ tmp["나이"] == i ]["value"])
        else: # 100세 이상
            sum10 = sum10 + int(tmp[ tmp["나이"] == i ]["value"])
 
    sum0_list.append(sum0)
    sum1_list.append(sum1)
    sum2_list.append(sum2)
    sum3_list.append(sum3)
    sum4_list.append(sum4)
    sum5_list.append(sum5)
    sum6_list.append(sum6)
    sum7_list.append(sum7)
    sum8_list.append(sum8)
    sum9_list.append(sum9)
    sum10_list.append(sum10)
```

```python
pop_cent = pd.DataFrame({
    '행정구역':state_list,
    '유아':sum0_list,
    '10대':sum1_list,
    '20대':sum2_list,
    '30대':sum3_list,
    '40대':sum4_list,
    '50대':sum5_list,
    '60대':sum6_list,
    '70대':sum7_list,
    '80대':sum8_list,
    '90대':sum9_list,
    '100세 이상':sum10_list,
})
pop_cent.head(2)
```

```python
state_list = test['행정구역'].unique()
# sum0_list = list()
sum1_list = list()
sum2_list = list()
sum3_list = list()
sum4_list = list()
sum5_list = list()
sum6_list = list()
sum7_list = list()

for one in state_list:
    tmp = test[ test["행정구역"] == one ]
#     sum0 = 0
    sum1 = 0
    sum2 = 0
    sum3 = 0
    sum4 = 0
    sum5 = 0
    sum6 = 0
    sum7 = 0
    
    for i in tmp.나이:
        if re.match(age0, i):
            continue
        elif re.match(age10, i):
            sum1 = sum1 + int(tmp[ tmp["나이"] == i ]["value"])
        elif re.match(age20, i):
            sum2 = sum2 + int(tmp[ tmp["나이"] == i ]["value"])
        elif re.match(age30, i):
            sum3 = sum3 + int(tmp[ tmp["나이"] == i ]["value"])
        elif re.match(age40, i):
            sum4 = sum4 + int(tmp[ tmp["나이"] == i ]["value"])
        elif re.match(age50, i):
            sum5 = sum5 + int(tmp[ tmp["나이"] == i ]["value"])
        elif re.match(age60, i):
            sum6 = sum6 + int(tmp[ tmp["나이"] == i ]["value"])
        else:
            sum7 = sum7 + int(tmp[ tmp["나이"] == i ]["value"])
 
#     sum0_list.append(sum0)
    sum1_list.append(sum1)
    sum2_list.append(sum2)
    sum3_list.append(sum3)
    sum4_list.append(sum4)
    sum5_list.append(sum5)
    sum6_list.append(sum6)
    sum7_list.append(sum7)
```
