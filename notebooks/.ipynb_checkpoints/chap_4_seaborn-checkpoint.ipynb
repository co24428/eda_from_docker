```python
tips = sns.load_dataset("tips") 
print(tips.head())
print(type(tips)) # <class 'pandas.core.frame.DataFrame'>
```

```python
print(tips.shape) # 244 row x 7 column
```

```python
fig = plt.figure() 
axes1 = fig.add_subplot(1, 1, 1)
```

```python
# 히스토그램 => 각각의 row값들이 연속적 ( 수치 값, ...)
# 막대그래프 => 각각의 row값들이 비연속적 ( 카테고리, ...)
axes1.hist(tips['total_bill'], bins=10) 
axes1.set_title('Histogram of Total Bill')
axes1.set_xlabel('Frequency') 
axes1.set_ylabel('Total Bill') 

fig
```

```python
scatter_plot = plt.figure() 
axes1 = scatter_plot.add_subplot(1, 1, 1) 
axes1.scatter(tips['total_bill'], tips['tip']) 
axes1.set_title('Scatterplot of Total Bill vs Tip') 
axes1.set_xlabel('Total Bill') 
axes1.set_ylabel('Tip')
```

```python
boxplot = plt.figure() 
axes1 = boxplot.add_subplot(1, 1, 1) 

axes1.boxplot( 
    [tips[tips['sex'] == 'Female']['tip'], 
     tips[tips['sex'] == 'Male']['tip']], 
    labels=['Female', 'Male'])


axes1.set_xlabel('Sex') 
axes1.set_ylabel('Tip') 
axes1.set_title('Boxplot of Tips by Sex')
```

## 다변량 데이터로 다변량 그래프 그리기 - 산점도 그래프(89쪽)

```python
def recode_sex(sex):
    if sex == 'Female':
        return 0 
    else:
        return 1
```

```python
tips['sex_color'] = tips['sex'].apply(recode_sex)
```

```python
scatter_plot = plt.figure() 
axes1 = scatter_plot.add_subplot(1, 1, 1) 
axes1.scatter(
    x=tips['total_bill'], 
    y=tips['tip'],    
    s=tips['size'] * 10,    
    c=tips['sex_color'],
    alpha=0.5)

axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size') 
axes1.set_xlabel('Total Bill') 
axes1.set_ylabel('Tip')
```

## 단변량 그래프 그리기 - 히스토그램(91쪽)

```python
import seaborn as sns

tips = sns.load_dataset("tips")
```

```python
# distplot 히스토그램과 선그래프를 동시에
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'])  
ax.set_title('Total Bill Histogram with Density Plot')
```

```python
# distplot => hist 옵션을 통해 히스토그램 제거
ax = plt.subplots() 
ax = sns.distplot(tips['total_bill'], hist=False) 
ax.set_title('Total Bill Density') 
ax.set_xlabel('Total Bill') 
ax.set_ylabel('Unit Probability')
```

```python
hist_den_rug, ax = plt.subplots() 
ax = sns.distplot(tips['total_bill'], rug=True) # rugplot => 실제 데이터의 위치
ax.set_title('Total Bill Histogram with Density and Rug Plot') 
ax.set_xlabel('Total Bill')
```

```python
# countplot
ax = plt.subplots() 
ax = sns.countplot('day', data=tips) 
ax.set_title('Count of days') 
ax.set_xlabel('Day of the Week') 
ax.set_ylabel('Frequency')
```

## 다양한 종류의 이변량 그래프 그리기(95쪽)

### 1. seaborn 라이브러리로 산점도 그래프 그리기

```python
# scatterplot
ax = plt.subplots() 
ax = sns.regplot(x='total_bill', y='tip', data=tips) 
ax.set_title('Scatterplot of Total Bill and Tip') 
ax.set_xlabel('Total Bill') 
ax.set_ylabel('Tip')
```

```python
# scatterplot => 중앙 선 없이 (fit_reg)
ax = plt.subplots() 
ax = sns.regplot(x='total_bill', y='tip', data=tips, fit_reg=False) 
ax.set_title('Scatterplot of Total Bill and Tip') 
ax.set_xlabel('Total Bill') 
ax.set_ylabel('Tip')
```

```python
# jointplot => scatter + each row & column's histogram
joint = sns.jointplot(x='total_bill', y='tip', data=tips) 
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

```python
# jointplot => kind -> 스캐터플롯 형태
hexbin = sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex") 
hexbin.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
hexbin.fig.suptitle('Hexbin Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

### 4. 이차원 밀집도 그리기

```python
# kde ( Kernel Density Plot )
kde, ax = plt.subplots() 
ax = sns.kdeplot(data=tips['total_bill'],
                 data2=tips['tip'], 
                 shade=True)
ax.set_title('Kernel Density Plot of Total Bill and Tip') 
ax.set_xlabel('Total Bill') 
ax.set_ylabel('Tip')
```

### 5. 바 그래프 그리기

```python
# barplot => y축은 x축을 알아서 조인하여 평균값?
ax = plt.subplots() 
ax = sns.barplot(x='time', y='total_bill', data=tips) 
ax.set_title('Bar plot of average total bill for time of day') 
ax.set_xlabel('Time of day') 
ax.set_ylabel('Average total bill')
```

### 6. 박스 그래프 그리기

```python
# boxplot
# ---- OUTLIER를 제외한 maximum
# 75% 선
# median => 50%에 위치한 데이터
# 25% 선
# ---- OUTLIER를 제외한 minimum
ax = plt.subplots() 
ax = sns.boxplot(x='time', y='total_bill', data=tips) 
ax.set_title('Boxplot of total bill by time of day') 
ax.set_xlabel('Time of day') 
ax.set_ylabel('Total Bill')
```

```python
ax = plt.subplots() 
ax = sns.violinplot(x='time', y='total_bill', data=tips) 
ax.set_title('Violin plot of total bill by time of day') 
ax.set_xlabel('Time of day') 
ax.set_ylabel('Total Bill')
```

### 8. 관계 그래프 그리기

```python
fig = sns.pairplot(tips)
```

```python
# fit_reg => 중앙선
# hue => 점 구분해주는 것
# scatter_kws => scatterplot의 설정잡기 / 's'- size (number값이 들어가야 한다)
scatter = sns.lmplot(x='total_bill', y='tip', data=tips, fit_reg=False, hue='sex', scatter_kws={'s': tips['size']*10}) 
plt.show()
```
